```python
# since we are told that we cannot use the threshold function, we would need to implement our own.
# Otsu's algorithm will only find us the value of the threshold, we have to binarize the
# image ourselves. So implementing a simply binary threshold function first

def thresholdBinary_ForLoops (img, boundary):
    '''takes an image, converts to grayscale if needed, sets pixels below threshold boundary to 0 and above to 255
    Uses nested for loops to traverse pixels'''
    if boundary <0 or boundary>255:
        print("Invalid threshold boundary")
        return -1
    if len(img.shape) == 3: # trying to convert an image that is already grayscale gives an error,
                            # so first check if it is coloured
        imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        imggray = img.copy()
#     print(imggray.shape)
    Y, X = imggray.shape
    for y in range(Y):
        for x in range(X):
            pixel = imggray[y][x]
            if pixel <= boundary:
                imggray[y][x] = 0
            else:
                imggray[y][x] = 255
    
    return imggray
                
```

```python
## Above works, but trying to use Numpy vector operations instead of nested for loops, faster and elegant
def thresholdBinary(img, boundary):
    '''takes an image, converts to grayscale if needed, sets pixels below boundary to 0 and above to 255
    Uses Numpy vector operations'''
    if boundary <0 or boundary>255:
        print("Invalid threshold boundary")
        return -1
    if len(img.shape) == 3: # trying to convert an image that is already grayscale gives an error,
                            # so first check if it is coloured
        imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        imggray = img.copy()
#     print(imggray.shape)
    
    imggray[imggray<=boundary] = 0 #Numpy magic
    imggray[imggray>boundary] = 255 #did 1 instead of 255 here by mistake earlier. Resulted in image with just 0 & 1s
                                    #plt.imshow(...'gray') was still showing it correctly, but cv.imshow() was showing 
                                    #as black. Another way I found is to use cv.convertScaleAbs() later
    
    return imggray
        
```

```python
def Otsushi (img):
    '''tries to implement Otsu in a semi-efficient way. Attempting Numpy broadcast arithmetic
    rather than lots of nested for loops'''

    # convert to grayscale if needed
    if len(img.shape) == 3: # trying to convert an image that is already grayscale gives an error,
                            # so first check if it is coloured
        imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        imggray = img.copy()
    
    # Otsu doesn't care about 2D shapes in image, just the intensity values as a whole,
    # so flatten it first
    flattened = imggray.ravel()

    # we need the probability of a pixel being of a particular intensity. This is easiest to calculate from the
    # histogram, which already gives us a probability distribution function
    hist, edges = np.histogram(imggray, bins=256, range=(0, 255)) # if we don't specify the last argument, it will
    # make bins only from lowest intensity present in image to highest, which wouldn't cover full intensity space

    # our probabilities should sum up to 1, so we need to normalize
    # print("DEBUG histogram sum is", np.sum(hist))
    hist = hist/np.sum(hist)

    # set initial threshold to impossible value
    boundary = -1
    
    # will go with maximizing inter-class distance, 
    # since is easier to compute this than minimizing intra-class variance
    
    # set up initial maximum to impossible value
    maxDistance = -1
    
    # now iterate for all possible thresholds from 0-255 to find the boundary value that maximizes
    # inter-class distance
    for t in range(256):
        
        # total probability (weight) of pixels before and after the current boundary
        weight1 = np.sum(hist[:t]) # till boundary
        weight2 = np.sum(hist[t:]) # after boundary

        # total of intensity values of all pixels till threshold boundary
        area1 = np.sum(np.arange(t) * hist[:t]) #arange() gives us the intensity value at each bin (x-axis), 
                    # hist contains how many pixels are there (y-axis)(now normalized, so would be a probability).
                    # Effectively, it gives us the total (statistical) expectation of intensity values
                    # We can also say that we are computing the area under the histogram curve
        mean1 = area1/weight1 # mean intensity of pixels before the threshold
            
        # repeat same process on other side of threshold boundary
        area2 = np.sum(np.arange(t,256) * hist[t:])
        mean2 = area2/weight2
        
        # now find inter-class variance.
        distance = weight1 * weight2 * np.power((mean2-mean1),2)
        
        # update max distance if needed, and set new threshold boundary at that point
        if maxDistance < distance:
            maxDistance = distance
            boundary = t
    
    return boundary

# above algorithm is working, but can be improved by making process incremental (instead of calculating everything
# in every iteration of the for loop, we can just add/subtract the new pixels that come in/go out when threshold moves).
# Will investigate again if time permits
```

```python
def Otsu (filepath):
    img = cv2.imread(filepath)
    threshval = Otsushi(img) # converting to grayscale etc would be done by this function
    print("The threshold value is", threshval)
    imgresult = thresholdBinary(img, threshval)
    return imgresult
```

**Solution: We can see above that for the image c1.jpg, the optimal threshold along the value dimension is 220 (according to Otsu)**

```python
#now combine all of the above, start with full image and binarise along value

def threshold_HSV_Value (filepath):
    img = cv2.imread(filepath)
    img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    V = img_HSV.copy()
    V[:, :, 0] = 0 # set H to 0
    V[:, :, 1] = 0 # set S to 0
    V_RGB =  cv2.cvtColor(V,cv2.COLOR_HSV2RGB) # convert back to RGB 
#     plt.imshow(V_RGB)
    imgValue = cv2.cvtColor(V_RGB, cv2.COLOR_BGR2GRAY)
    threshval_Value = Otsushi(imgValue)
#     print(threshval_Value)
    imgresult = thresholdBinary(imgValue, threshval_Value)
    return imgresult, threshval_Value
```

### Task 2.1: Image Cleaning, Median filter

There are some small cells and/or noise in the picture. We should use a low-pass filter to reduce noise. A median filter is not as computation intensive as Gaussian Blur, and works reasonably well.

In this section, will be using the segmented output from the HSV value thresholder in the previous section, as it looks better than the Otsu threshold output on RGB. 


```python
# had made this median filter for Lab 1 (Week 2). Reusing the same, since we are not allowed to use OpenCV medianBlur

def MedianFilter(img, ksize):
    out_img = np.zeros(img.shape)
    img = np.pad(img, pad_width=ksize//2, mode='constant', constant_values=0) # zero-padding
    X,Y = img.shape
    for i in range(0, X - (ksize-1)):
        for j in range(0, Y - (ksize-1)):
            out_img [i,j] = np.median(img[i:i+ksize, j:j+ksize]) # find the median and put in the output pixel
        
    return np.uint8(out_img) # as seen in Question 1 above, not using cast causes problems with cv2.imshow()
```

### Task 2.2 Connected components counting

After a lot of reading and research, found that before counting components, we have to essentially do a flood-fill on each local cell area with a different colour. 

Creating a class for this so that large images don't need to be passed back and forth between functions.

```python
class ColorMe:
    '''implements flood-fill methods and counts components.'''

    # was opening file from scratch initially. Then realised if do that,
    # then would need to do grayscale, Otsu, HSV threshold, median filter etc 
    # all inside this class, would make it unnecessarily long and harder to
    # understand. Would also repeat a lot of work that has already been done
    # so taking img object in constructor now rather than filepath
    #     def __init__ (self, filepath): 
    #         img = cv2.imread(filepath)

    def __init__ (self, img, colorToChange = 0):
        '''colorToChange is the color we plan to change. 0 (black) in our question'''            
        self.img = img
        self.height = img.shape[0]
        self.width = img.shape[1]
        self.colorToChange = colorToChange
        self.colorTicker = 1 # first cell area would be colored 1, then this
        # would be incremented to 2 for coloring next cell and so on
        
    def giveDims(self):
        # sometimes imread fails silently if path or filetype is not correct,
        # this gives dimensions, helps check if image was loaded
        print(self.height, self.width)
    
    def floodFill (self, colorNew, i, j): #this is Numpy, so i & j are y&x, not x&y
        if self.img[i,j] == colorNew:
            return
        else:
            colorOld = self.img[i,j]
            self.fillRec (i, j, colorOld, colorNew) # calling recursive function
    
    def fillRec (self, i, j, colorOld, colorNew):
        '''Recursively fills out area of old color with new color'''
        
        # base case recursion tests: we have reached bounds of image, 
        # or a boundary pixel of some other color (maybe one filled by us already)
        if i<0 or i>=self.height or j<0 or j>=self.width or self.img[i,j]!=colorOld:
            return
        
        # assign new color to pixel
        self.img[i,j] = colorNew
        
        # and then do it recursively on 4-neighborhood
        self.fillRec (i-1, j, colorOld, colorNew) #above
        self.fillRec (i+1, j, colorOld, colorNew) #below
        self.fillRec (i, j-1, colorOld, colorNew) #left
        self.fillRec (i, j+1, colorOld, colorNew) #right

    def getImage(self):
        return self.img
    
    def fillCells(self):
        for i in range(self.height):
            for j in range(self.width):
                # if cell of old color (0) found, fill it with a new color
                # and increment the color
                if self.img[i,j] == self.colorToChange:
                    self.floodFill(self.colorTicker, i, j)
                    self.colorTicker += 1
    
    def getCurrentColor(self): # returns the color number we are currently at
        # would have incremented based on the number of areas we have filled
        return self.colorTicker
        
    def getColorPixelCounts(self):
        colors, counts = np.unique(self.img, return_counts=True)
        # counts gives us pixel counts (area) of various colors
        return colors, counts

# end class
```

On visual inspection of the above list, we can say that the best values for normal RBC cell size lie in the middle region of this sorted list. Values at the beginning are noise or mostly cut-off edge cells, while big values at the end are overlapping double cells.

There is no good automated way to calculate the min_area that would remain valid across images. We can try the mean, but that would be affected by number and size of cut-off cells at one end, and overlapping cells at the other end. These would differ from picture to picture. Using the median will find a normal cell, and not be able to account for cut-off cells. 

So using domain knowledge and visual inspection of this data and the source image, we can see that there is a big jump from area 32 to 424 as we move from the 10th to the 11th element in the list. This indicates a substantial change in size. I am **setting 400 as the min_area**.

For overlap, we can see that at the higher end of the scale, values increase from ~2000 (1939) suddenly to ~3000 (2909). This almost certainly indicates overlapping cells. We can also look at the filtered image and see that there are indeed 3 overlapping cells. I am **setting 2500 as the overlap_area**, which is comfortably above the 1939 of the previous largest single cell.

**Solution: Min and overlap areas** (reasoning discussed in whole section above)

```python
min_area = 400
overlap_area = 2500
```

```python
def CountCells (pixelareas, min_area, overlap_area):
    '''takes a list containing pixel areas and counts total cells,
    based on min area and overlap area'''
    areas = np.array(pixelareas) #so can use faster vector operations
    normal_cell_count = np.sum(areas>min_area)
#     print(normal_cell_count)
    # we have counted all cells above the min_area, but now we need to 
    # double count those above the overlap area, since they are two cells
    double_cell_count = np.sum(areas>overlap_area)
#     print(double_cell_count) # e.g. for c1.jpg this gives 3, as expected
    return (normal_cell_count + double_cell_count)
```

```python
c1_cellCount = CountCells(counts, min_area, overlap_area)
c1_cellCount
```

The above cell counting function can be moved inside the class, but I want to keep the explanation and reasoning of why I chose these values. So leaving it here outside the class.

**Solution: the cell count for image c1.jpg is 57** (according to my approach, as calculated and explained above)
